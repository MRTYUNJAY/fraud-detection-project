```python
from google.colab import files
uploaded = files.upload()

import zipfile
with zipfile.ZipFile('fraudTest.csv.zip', 'r') as zip_ref:
    zip_ref.extractall()

import pandas as pd
df = pd.read_csv('fraudTest.csv')
df.head(3)
```

```python
print(df.isnull().sum())
#no null values in our dataset
```

```python
df["FUll Name"] = df["first"] + " " + df["last"]
df["address"] = df["street"] + " " + df["city"] + " " + df["state"] + " " + df["zip"].astype(str) + " " + df["lat"].astype(str) + " " + df["long"].astype(str) + " " + df["city_pop"].astype(str)
df.drop(['first', 'last', 'street', 'city', 'state', 'zip', 'lat', 'long', 'city_pop','gender','dob','trans_num'], axis=1, inplace=True)
df.head(3)
```

```python
df.drop(['Unnamed: 0','trans_date_trans_time','merch_lat','merch_long'], axis=1, inplace=True)
df.head(3)
```

```python
X=df.drop(columns='is_fraud')
y=df['is_fraud']
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

# X: Features, y: Labels
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42)
print("Train set fraud ratio:", sum(y_train) / len(y_train))
print("Test set fraud ratio:", sum(y_test) / len(y_test))
```

```python
import numpy as np
num_columns=X_train.select_dtypes(include=np.number).columns
cat_columns=X_train.select_dtypes(include='object').columns
print(num_columns)
print(cat_columns)
```

```python
#I would be using random forest so no scaling required
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
encoder.fit(X_train[cat_columns])
X_train_cat_encoded=encoder.transform(X_train[cat_columns])
X_test_cat_encoded=encoder.transform(X_test[cat_columns])
df_cat_encoded=pd.DataFrame(X_train_cat_encoded,columns=encoder.get_feature_names_out())
df_cat_encoded
```

```python
X_train_num_scaled=X_train[num_columns]
X_test_num_scaled=X_test[num_columns]
X_train_final=pd.concat([X_train_num_scaled,df_cat_encoded],axis=1)
X_test_final=pd.concat([X_test_num_scaled,pd.DataFrame(X_test_cat_encoded,columns=encoder.get_feature_names_out())],axis=1)
X_train_final
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
model=RandomForestClassifier(n_estimators=100,n_jobs=-1,random_state=42)
model.fit(X_train_final,y_train)
y_pred=model.predict(X_test_final)
model.score(X_test_final,y_test),model.score(X_train_final,y_train)
```
